# Locators

Ways to identify one or more specific elements in the DOM. 

"DOM" stands for "Document Object Model" in web programming. It is a data structure that the browser uses to represent and interact with elements on a web page. The DOM allows elements on a web page to be accessed, manipulated, and updated through JavaScript code.


https://www.selenium.dev/documentation/webdriver/elements/locators/

Selenium provides support for these 8 traditional location strategies in WebDriver:


```python
url = 'https://demo.guru99.com/test/login.html'

driver.get(url)
```

| Locator | Description  |
|---|---|
| class name | Locates elements whose class name contains the search value (compound class names are not permitted)  |
| id | Locates elements whose ID attribute matches the search value  |
| name | Locates elements whose NAME attribute matches the search value  |
| link text | Locates anchor elements whose visible text matches the search value  |
| partial link text | Locates anchor elements whose visible text contains the search value. If multiple elements are matching, only the first one will be selected.  |
| tag name | Locates elements whose tag name matches the search value  |
| xpath | Locates elements matching an XPath expression |

### Find element by `Xpath`

Xpath stands for "XML Path Language". It is a language used to locate and retrieve elements in an XML document. Here are some examples of Xpath shortcuts:

+ //: Represents all elements in the XML document.
+ /: Used to access a child of an element.
+ @: Used to access an attribute of an element.


![](https://www.guru99.com/images/3-2016/032816_0758_XPathinSele1.png)

> The XPath could be absolute xpath, which is created from the root of the document.

Example - /html/form/input[1].

> Or the xpath could be relative. 

Example- //input[@name=‘fname’]

```python
/html/body/div[4]/div/div/div[2]/div/div/div/form/div/div[1]/input
```

```python
//*[@id="email"]
```

```python
# using absolute xPath:
driver.find_elements(by=By.XPATH, value='/html/body/div[4]/div/div/div[2]/div/div/div/form/div/div[1]/input')
```

```python
driver.find_elements(by=By.XPATH, value="//input[@data-validate='isEmail']")
```

## Get Element's Information

### is displayed: `.is_displayed()`

This method is used to check if the connected Element is displayed on a webpage. Returns a Boolean value, True if the connected element is displayed in the current browsing context else returns false.

```python
# Navigate to the url
driver.get("https://www.selenium.dev/selenium/web/inputs.html")
```

### Is Selected `.is_selected()`
This method determines if the referenced Element is Selected or not. This method is widely used on `Check boxes, radio buttons, input elements, and option elements`.

Returns a boolean value, True if referenced element is selected in the current browsing context else returns false.

```python
# Returns true if element is checked else returns false
tick_elm = driver.find_elements(by = By.XPATH, value='//input[@name="checkbox_input"]')
tick_elm[0].is_selected()
```

### Tag Name `.tag_name`
It is used to fetch the TagName of the referenced Element which has the focus in the current browsing context.

```python
# Returns TagName of the element
emailInput_elm = driver.find_element(By.NAME, "email_input")
emailInput_elm
```

```python
emailInput_elm.tag_name
```

### Text Content `.text` or `.get_attribute('textContent')`
Retrieves the rendered text of the specified element.

```python
elm = driver.find_element(by=By.XPATH, value='/html/body')
elm.text
```

```python
elm.get_attribute('textContent')
```

### Fetching Attributes or Properties `.get_attribute("value")`

Fetches the run time value associated with a DOM attribute. It returns the data associated with the DOM attribute or property of the element.

```python
# Identify the email text box
emailInput_elm = driver.find_element(by=By.XPATH, value='/html/body/form/input[4]')
emailInput_elm.get_attribute('value')
```

## Website Interactions:


### Click: `.click()`

```python
# Click on the element 
color_elm = driver.find_element(by = By.XPATH, value='/html/body/form/input[13]')
color_elm.click()
```

### Clear & Send keys `.clear()` and `send_keys()`
